```python
import matplotlib.pylab as plot
import numpy as np
```

```python
def plot_circle(rad,cx,cy):
    radrad=np.radians(rad)
    r=np.radians(np.arange(0.,365,5))
    x=np.radians(cx)+radrad*np.cos(r)
    y=radrad*np.sin(r)-np.radians(cy)
    plot.plot(x,y)
    
def plot_box(rx,ry,cx,cy):
    xarr=np.arange(cx-rx/2.,cx+rx/2.,1)
    nx=len(xarr)
    yarr=np.arange(cy-ry/2.,cy+ry/2.,1)
    ny=len(yarr)
    x=np.append(np.append(xarr,np.repeat(cx+rx/2.,ny)),np.append(xarr[::-1],np.repeat(cx-rx/2.,ny)))
    y=np.append(np.append(np.repeat(cy-ry/2.,nx),yarr),np.append(np.repeat(cy+ry/2.,nx),yarr[::-1]))
    x=np.radians(x)
    y=np.radians(y)
    plot.plot(x,y)
    
def random_orbit(inclination):
    r=np.radians(np.arange(0.,365,5))
    x=r+np.random.rand()*2*np.pi-2*np.pi
    y=np.radians(inclination)*np.sin(r)
    return np.degrees(x),np.degrees(y)
```

```python
m=plot.subplot(111,projection='aitoff')
#m=plot.figure()
m.axes.xaxis.set_ticklabels([])
m.grid(True)
```

```python
#move around FoVs with random sinusoidal pattern
xgbm,ygbm=random_orbit(26)
xbat,ybat=random_orbit(21)
xbc,ybc=random_orbit(55)
b=int(np.round(np.random.rand()*73)) #selecting random starting point
plot_circle(70,xgbm[b],ygbm[b]) #GBM
plot_box(90,50,xbat[b],ybat[b]) #BAT
plot_circle(60,xbc[b],ybc[b]) #BC
print xgbm[b],ygbm[b]
print xbat[b],ybat[b]
print xbc[b],ybc[b]
```
